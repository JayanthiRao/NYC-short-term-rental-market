```python
import numpy as np
import pandas as pd
```

```python
prices = pd.read_csv("prices.csv")
prices 
```

```python
xls = pd.ExcelFile('room_types.xlsx')
print(xls.sheet_names)
room_types = xls.parse(0)
room_types
```

```python
reviews = pd.read_table('reviews.tsv')
reviews
```

```python
prices['price']= prices['price'].str.replace(" dollars","")
prices['price'] = prices['price'].astype('int')
```

```python
prices.info()
```

```python
prices.describe()
```

```python
prices.sort_values('price')
```

```python
prices_new = prices[(prices['price']!=0)&(prices['price']!=7500)].copy()
prices_new
```

```python
prices_new.describe()
```

```python
avg_price = round(prices_new['price'].mean(),2)
avg_price
```

```python
prices_new['price_per_month'] = prices_new['price']*365/12

prices_new
```

```python
avg_price_per_month = prices_new['price_per_month'].mean()
avg_price_per_month
```

```python
room_types['room_type'].unique()
```

```python
room_types['room_type'] = room_types['room_type'].str.lower()
room_types['room_type'].unique()
```

```python
reviews['last_review'] = pd.to_datetime(reviews['last_review'], infer_datetime_format =True, errors ='coerce')
reviews
```

```python
room_and_price = prices_new.merge(room_types, on='listing_id', how ='outer')
room_and_price
```

```python
data_merged = room_and_price.merge(reviews, on ='listing_id', how ='outer')
data_merged
```

```python
data_merged.dropna(inplace=True)
data_merged
```

```python
data_merged.duplicated().sum()
```

```python
data_merged['borough'] = data_merged['nbhood_full'].str.partition(sep = ",",expand=True)[0]
boroughs = data_merged.groupby('borough')['price'].agg(['sum', 'mean', 'median', 'count'])
boroughs
```

```python
boroughs.round(2).sort_values('mean', ascending=False)
```

```python
label_names = ['Budget', 'Average', 'Expensive','Extravagant']
label_range = [0,69,175,350,np.inf]
data_merged['price_range'] = pd.cut(data_merged['price'], bins= label_range, labels =label_names)
data_merged.groupby(['borough','price_range'])['price_range'].count()
```
